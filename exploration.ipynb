```python
def get_close_events(df, query, dist):
    '''
    Gets events to query within dist
    '''
    lat_mask = np.abs(df.ActionGeo_Lat-query[0])<dist
    long_mask = np.abs(df.ActionGeo_Long-query[1])<dist
    cum_mask = lat_mask & long_mask
    filtered = df[cum_mask]
    return filtered
```

```python
def create_scores(df):
    '''
    Adds a column to the dataframe based on algorithm
    '''
    df['score'] = df['AvgTone']*df['NumMentions']
    return df
```

```python
def rank_events(df):
    '''
    Ranks events within the dataframe
    '''
    df = df.sort_values(by=['score'])
    return df
```

```python
def run_event_query(df, query, dist):
    f = get_close_events(df, query, dist)
    f_scored = create_scores(f)
    f_ranked = rank_events(f_scored)
    return f_ranked
```

```python
from sumy.parsers.html import HtmlParser
from sumy.summarizers.lsa import LsaSummarizer as Summarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
from sumy.nlp.tokenizers import Tokenizer

LANGUAGE = 'english'
SENTENCES_COUNT = 3
```

```python
def get_summary(url):
    try:
        parser = HtmlParser.from_url(url, Tokenizer(LANGUAGE))
        stemmer = Stemmer(LANGUAGE)

        summarizer = Summarizer(stemmer)
        summarizer.stop_words = get_stop_words(LANGUAGE)

        return ' '.join(map(str,summarizer(parser.document, SENTENCES_COUNT)))
    except:
        return None
```

```python
def get_full_query(df, query, num_results=5, dist=0.1):
    output = []
    event_list = run_event_query(df, query, dist)
    num_iters = num_results
    i = 0

    # keep track of unique URLs
    visited_urls = set()
    
    # iterate until end of event list or requisite number of results
    while num_iters and i < len(event_list):
        # get relevant row in dataframe
        event = event_list.iloc[i]

        i += 1
        if event.SOURCEURL in visited_urls:
            continue

        # see if url is valid for summary
        summary = get_summary(event.SOURCEURL)
        if not summary:
            continue
        
        # create result dictionary
        result = {}
        result['url'] = event.SOURCEURL
        result['text'] = summary
        result['score'] = event.score
        result['lat'] = str(event.ActionGeo_Lat)
        result['long'] = str(event.ActionGeo_Long)
        result['date'] = str(event.SQLDATE)
        
        # append result
        output.append(result)
        
        # iterate
        visited_urls.add(event.SOURCEURL)
        num_iters -= 1
    
    # output
    print (json.dumps(output))
```

```python
from sumy.parsers.html import HtmlParser
from sumy.summarizers.lsa import LsaSummarizer as Summarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
from sumy.nlp.tokenizers import Tokenizer
```

```python
import json
import nltk
from nltk.corpus import wordnet
```

```python
DATA_LIST = [
    'data/20190130.export.CSV',
    'data/20190131.export.CSV',
    'data/20190201.export.CSV',
]
```

```python
class DataBase():
    def __init__(self, DATA_LIST, SCORE_THRESHOLD=-100, LANGUAGE='english',
                 SENTENCES_COUNT=4, NUM_RESULTS=5):
        self.SCORE_THRESHOLD=SCORE_THRESHOLD
        self._initialize_data(DATA_LIST)
        self.LANGUAGE = LANGUAGE
        self.SENTENCES_COUNT = SENTENCES_COUNT
        self.NUM_RESULTS = NUM_RESULTS
        self.KEYWORDS = None
    
    def _initialize_data(self, data_list):
        '''
        Consolidates data into one file with score thresholds below given threshold
        '''
        print('\n **INITIALIZING DATA**')

        header_string="GLOBALEVENTID	SQLDATE	MonthYear	Year	FractionDate	Actor1Code	Actor1Name	Actor1CountryCode	Actor1KnownGroupCode	Actor1EthnicCode	Actor1Religion1Code	Actor1Religion2Code	Actor1Type1Code	Actor1Type2Code	Actor1Type3Code	Actor2Code	Actor2Name	Actor2CountryCode	Actor2KnownGroupCode	Actor2EthnicCode	Actor2Religion1Code	Actor2Religion2Code	Actor2Type1Code	Actor2Type2Code	Actor2Type3Code	IsRootEvent	EventCode	EventBaseCode	EventRootCode	QuadClass	GoldsteinScale	NumMentions	NumSources	NumArticles	AvgTone	Actor1Geo_Type	Actor1Geo_FullName	Actor1Geo_CountryCode	Actor1Geo_ADM1Code	Actor1Geo_Lat	Actor1Geo_Long	Actor1Geo_FeatureID	Actor2Geo_Type	Actor2Geo_FullName	Actor2Geo_CountryCode	Actor2Geo_ADM1Code	Actor2Geo_Lat	Actor2Geo_Long	Actor2Geo_FeatureID	ActionGeo_Type	ActionGeo_FullName	ActionGeo_CountryCode	ActionGeo_ADM1Code	ActionGeo_Lat	ActionGeo_Long	ActionGeo_FeatureID	DATEADDED	SOURCEURL"
        headers = header_string.split('\t')

        self.DATA = pd.DataFrame(columns=headers)
        
        for df_fp in data_list:
            print('READING {data}'.format(data=df_fp))

            # read in dataframe
            df = pd.read_csv(df_fp, sep='\t', header=None)
            df.columns = headers
            
            # create score column
            df = self._create_scores(df)
            
            # create mask for threshold
            mask = df['score'] < self.SCORE_THRESHOLD
    
            # append masked dataframe to self.DATA
            self.DATA = self.DATA.append(df[mask])
    
        # sort data by score
        self.DATA = self._rank_events(self.DATA)
        
        print('INITIALIZED DATA')

    def query(self, query_list, out_path='query_result.json'):
        '''
        Runs a query across all files in the DATA_LIST
        
        query_list: a list of tuples (location_name, location_lat, location_long, dist, keywords)

        Outputs a dictionary with the date and significant events for each location
        '''
        print('\n **RUNNING QUERY**')
        self.output = {}
        
        # create dictionary with location as keys
        for query in query_list:
            self.output[query[0]] = self._get_query_events(self.DATA, query)
    
        # return jsonified output
        with open(out_path, 'w') as outfile:
            json.dump(self.output, outfile)
            print('DUMPED TO {outfile}'.format(outfile=out_path))
        print('FINISHED QUERY')

    def _get_query_events(self, df, query):
        '''
        Updates self.output with events from the specified filepath
        '''
        location = query[0]
        print('QUERYING: {location}'.format(location=location))
        lat, long, dist = query[1], query[2], query[3]
        keywords = self._get_keyword_synonyms(query[4])
        return self._get_relevant_event_details(df, lat, long, dist, keywords)

    def _get_keyword_synonyms(self, KEYWORD_LIST):
        '''
        utilizes ntlk to create a set of synonyms
        '''
        keyword_list = set()
        for keyword in KEYWORD_LIST:
            # add initial keyword
            keyword_list.add(keyword)
    
            # get synonyms
            for syn in wordnet.synsets(keyword)[:3]:
                for l in syn.lemmas():
                    keyword_list.add(l.name())
        return keyword_list

    def _get_relevant_event_details(self, df, lat, long, dist, keywords):
        '''
        Gets relevant events (relevance = close & associated)
        '''        
        # get close events
        close_events = self._get_close_events(df, lat, long, dist)

        return self._get_associated_event_details(close_events, keywords)

    def _get_summary(self, url):
        '''
        Gets summary from the url
        '''
        try:
            parser = HtmlParser.from_url(url, Tokenizer(self.LANGUAGE))
            stemmer = Stemmer(self.LANGUAGE)

            summarizer = Summarizer(stemmer)
            summarizer.stop_words = get_stop_words(self.LANGUAGE)

            return ' '.join(map(str,summarizer(parser.document, self.SENTENCES_COUNT)))
        
        except:
            return None
    
    def _get_associated_event_details(self, df, keywords):
        '''
        Gets self.NUM_RESULTS events that have keywords
        and creates a description object
        '''
        output = []

        # get associated events
        num_iters = self.NUM_RESULTS
        i = 0

        # keep track of unique URLs
        visited_urls = set()

        # iterate until end of event list or requisite number of results
        while num_iters and i < len(df):
            # get relevant row in dataframe
            event = df.iloc[i]

            i += 1
            
            # check if URL has been visited before
            if event.SOURCEURL in visited_urls:
                continue

            # see if url is valid for summary
            summary = self._get_summary(event.SOURCEURL)
            if not summary:
                continue
            
            # convert summary into word set
            summary_set = set(summary.split())

            # see if there's an overlap between the summary
            if not len(keywords.intersection(summary_set)):
                continue

            # create result dictionary
            result = {}
            result['url'] = event.SOURCEURL
            result['text'] = summary
            result['score'] = event.score
            result['lat'] = str(event.ActionGeo_Lat)
            result['long'] = str(event.ActionGeo_Long)
            result['date'] = str(event.SQLDATE)

            # append result
            output.append(result)

            # iterate
            visited_urls.add(event.SOURCEURL)
            num_iters -= 1
        return output

    def _get_close_events(self, df, lat, long, dist):
        '''
        Gets events to query within dist
        '''
        lat_mask = np.abs(df.ActionGeo_Lat-lat)<dist
        long_mask = np.abs(df.ActionGeo_Long-long)<dist
        cum_mask = lat_mask & long_mask
        filtered = df[cum_mask]
        return filtered

    def _create_scores(self, df):
        '''
        Adds a column to the dataframe based on algorithm
        '''
        df['score'] = df['AvgTone']*df['NumMentions']
        return df

    def _rank_events(self, df):
        '''
        Ranks events within the dataframe
        '''
        df = df.sort_values(by=['score'])
        return df
    
    
        
```
